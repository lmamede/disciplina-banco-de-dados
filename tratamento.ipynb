```python
df2 = pd.read_csv('games_details.csv')

df2.head()
```

```python
df2.describe()
```

```python
def treat(x):
    pos = -1

    sec = 0

    for i in range(len(x)):
        if x[i] == ':':
            pos = i
            break

    if pos!=-1:
        sec = int(x[pos+1:])

        minu = float(x[:pos])

        minu = int(minu)

        sec+=minu*60

    else:
        minu = int(x)
        minu = abs(minu)

        sec = minu*60
    
    return sec
    
```

```python
nicks = {}
minutes = {}

for idx in df2.index:
    id = df2['PLAYER_ID'][idx]
    nick = df2['NICKNAME'][idx]
    minu = df2['MIN'][idx]

    if id in nicks:
        if pd.isna(nicks[id]) and pd.isna(nick)==0:
            nicks[id] = nick
    else:
        nicks[id] = nick
    
    if id in minutes:
        if pd.isna(minu)==0:
            minutes[id] += treat(minu)
    else:
        if pd.isna(minu)==0:
            minutes[id] = treat(minu)
        else:
            minutes[id] = 0         
```

```python
nomes = {}
```

```python
for idx in df3.index:
    id = df3['PLAYER_ID'][idx]
    name = df3['PLAYER_NAME'][idx]
    
    if id in nomes:
        if name != nomes[id]:
            print(name, nomes[id])

    nomes[id] = name
```

```python
ids = list(nomes.keys())

nomes = list(nomes.values())
```

```python
data = {'PLAYER_ID': ids, 'PLAYER_NAME':nomes}

df = pd.DataFrame(data)
```

```python
def ni(x):
    if x in nicks:
        return nicks[x]
    else:
        return pd.NA
```

```python
def sec(x):
    if x in minutes:
        return minutes[x]
    else:
        return 0
```

```python
df.head()
```

```python
df['NICKNAME'] = df['PLAYER_ID'].apply(ni)
```

```python
df['SECONDS_PLAYED'] = df['PLAYER_ID'].apply(sec)
```

```python
df.head()
```

```python
columns_player = ['PLAYER_ID', 'PLAYER_NAME', 'NICKNAME', 'SECONDS_PLAYED']
```

```python
with open('players.sql', 'w') as writer:
    for id in df.index:

        writer.write('INSERT INTO Player VALUES (')
        
        for i in range(len(columns_player)):
            column = columns_player[i]

            value = df[column][id]

            if type(value)==str:
                writer.write('\'')

                for c in range(len(value)):
                    if value[c] == '\'':
                        writer.write("''")
                    else:
                        writer.write(value[c])
                    
                
                writer.write('\'')
            else:
                if pd.isna(value):
                    writer.write('NULL')
                elif type(value)==int:
                    value = int(value)
                    writer.write(str(value))
                else:
                    writer.write(str(value))
            

            if i == len(columns_player)-1:
                writer.write(')')
            else:
                writer.write(', ')

    
        writer.write(';\n')
```
